```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 =pd.read_csv('rawdata.csv')  #reading the dataset and printing it.
```

```python
df_1['Frailty'] = df_1['Frailty'].map({'N': 0, 'Y': 1})
df_1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the graph background
sns.set(style="dark")

plt.figure(figsize=(15, 5))

# Plotting the correlation between Age and Frailty
plt.subplot(1, 3, 1)
sns.regplot(x='Age', y='Frailty', data=df_1, logistic=True, ci=None)
plt.title('Age vs Frailty')

# Plotting the correlation between Weight and Frailty
plt.subplot(1, 3, 2)
sns.regplot(x='Weight', y='Frailty', data=df_1, logistic=True, ci=None)
plt.title('Weight vs Frailty')

# Plotting the correlation between Height and Frailty
plt.subplot(1, 3, 3)
sns.regplot(x='Height', y='Frailty', data=df_1, logistic=True, ci=None)
plt.title('Height vs Frailty')
plt.show()
```
