```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2=pd.read_csv('/content/StudentsPerformance.csv')
```

```python
# Visualization the graphs
sns.set(style='darkgrid')


# 1. Gender vs Average Scores
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df_2, palette='Set2')
plt.title('Gender vs Math Scores')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.tight_layout()
plt.show()

# 2. Lunch Type vs Math Scores
plt.figure(figsize=(8, 6))
sns.barplot(x='lunch', y='math score', data=df_2, palette='pastel')
plt.title('Lunch Type vs Math Scores')
plt.xlabel('Lunch Type')
plt.ylabel('Average Math Score')
plt.tight_layout()
plt.show()

# 3. Reading and Writing Scores Correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score', data=df_2, color='purple')
plt.title('Reading vs Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()

# 4. Average Scores by Race/Ethnicity
plt.figure(figsize=(8, 6))
race_avg_scores = df_2.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()
race_avg_scores.plot(kind='bar', figsize=(8, 6))
plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. Count Plot: Test Preparation Course Completion
plt.figure(figsize=(8, 6))
sns.countplot(x='test preparation course', data=df_2, palette='Blues')
plt.title('Test Preparation Course Completion Status')
plt.xlabel('Course Completion Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 6. Heatmap of Scores
plt.figure(figsize=(10, 8))
scores_correlation = df_2[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(scores_correlation, annot=True, cmap='viridis')
plt.title('Correlation Heatmap of Scores')
plt.tight_layout()
plt.show()
```
